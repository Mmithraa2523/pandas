```python
import pandas as pd
df = pd.read_csv("bronze\HCLTECH.csv")
print(df.head(5))
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d-%m-%Y')
df.set_index('Date', inplace=True)
```

```python
column = ["Adj Close","Close","High","Low","Open","Volume"]
for col in column:
    df[col]= df[col].round().astype('int')
```

```python
df = df[df['Volume'] != 0]
```

```python
df.to_csv("silver\\hcltech.csv", index=True)
```

```python
df["Daily_return"] = ((df['Close']-df['Open'])/df['Open'])*100
```

```python
df['Price_Range'] = df['High'] - df['Low']
```

```python
df['Ma_5day'] = df['Close'].rolling(5).mean()
```

```python
df['Ma_20day'] = df['Close'].rolling(20).mean()
```

```python
df['Volatility_%'] = ((df['High'] - df['Low']) / df['Close']) * 100
```

```python
df['Volume_Change_%'] = df['Volume'].pct_change() * 100
```

```python
import numpy as np
df['Trend'] = np.where(df['Close'] > df['Ma_20day'], 'Uptrend', 'Downtrend')
```

```python
df.to_csv("gold\\hcltech.csv", index=True)
```
